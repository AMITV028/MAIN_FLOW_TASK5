```python
# How many people have heart disease and how many people don't have heart disease?
data.target.value_counts()
```

```python
data.target.value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Heart Disease')
plt.xlabel('0 = No Disease, 1 = Disease')
plt.ylabel('Amount')
plt.show()
```

```python
data.target.value_counts().plot(kind='pie', figsize=(8,6), colors=['red', 'green'], autopct='%1.1f%%')
plt.legend(["Disease","No Disease"])
plt.show();
```

1) 51.3% people have heart disease & 48.7% people have no Heart disease

```python
# 3)People of which sex has which type of chest pain most?
data.cp.value_counts()
```

```python
data.cp.value_counts().plot(kind='bar', color=['red', 'green', 'blue', 'yellow'])
plt.title('Chest Pain Type')
plt.xlabel('0 = Typical Angina, 1 = Atypical Angina, 2 = Non-Anginal Pain, 3 = Asymptomatic', fontsize = 7)
plt.xticks(rotation=0)
plt.ylabel('Amount')
plt.show()
```

```python
pd.crosstab(data.sex, data.cp)
```

```python
pd.crosstab(data.sex, data.cp).plot(kind='bar', figsize=(10,6), color=['red', 'green', 'blue', 'yellow'])
plt.title('Chest Pain Frequency for sex')
plt.xlabel('0 = Female, 1 = Male')
plt.xticks(rotation=0)
plt.ylabel('Amount')
plt.show()
```

Type 0 cp is more prone in Male where as in female type 0 and 2 are  nearly same.


```python
# 4) People with which chest pain are most prone to have Heart disease?
pd.crosstab(data.cp, data.target)
```

```python
pd.crosstab(data.cp, data.target).plot(kind='bar', figsize=(10,6), color=['green','orange'])
plt.title('Heart Disease Frequency for Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.xticks(rotation=0)
plt.ylabel('Amount')
plt.show()
```
